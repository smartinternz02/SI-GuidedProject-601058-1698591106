Independent and dependent variable split

```python
x=df.drop("quality",axis=1)
```

```python
y=df.quality
y = y.astype(int)
```

Train-test spliting

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

Modeling

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
model1 = DecisionTreeClassifier(max_depth=20,splitter='best',criterion='entropy')
model1.fit(x_train,y_train)

```

```python
p_pre1p=model1.predict(x_test)
p_pre1t=model1.predict(x_train)
```

```python
from sklearn.metrics import classification_report,accuracy_score
print(classification_report(y_true=y_test, y_pred=p_pre1p))
accuracy_score(y_test,p_pre1p)
```

```python
print(classification_report(y_true=y_train, y_pred=p_pre1t))
```

Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
model2 =RandomForestClassifier(n_estimators=500,criterion='entropy')
model2.fit(x_train,y_train)
```

```python
p_pre1p=model2.predict(x_test)
p_pre1t=model2.predict(x_train)
```

```python
print(classification_report(y_true=y_test, y_pred=p_pre1p))
```

Naive Bayesian Classification Model

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train,y_train)
```

```python
p_pre1p=gnb.predict(x_test)
p_pre1t=gnb.predict(x_train)
```

```python
from sklearn.metrics import classification_report,accuracy_score
print(classification_report(y_true=y_test, y_pred=p_pre1p))
```

```python
print(classification_report(y_true=y_train, y_pred=p_pre1t))
```
